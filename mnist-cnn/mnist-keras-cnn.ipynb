```python
import pandas as pd
import numpy as np
import keras
from keras.layers import Conv2D, Dense, Dropout, MaxPooling2D, Flatten
from keras.models import Sequential
import tensorflow as tf
tf.test.is_gpu_available(cuda_only=False, min_cuda_compute_capability=None)#if this is true, tensorflow is using gpu.
```

```python

from tensorflow.python.client import device_lib

def get_available_gpus():
    local_device_protos = device_lib.list_local_devices()
    return [x.name for x in local_device_protos if x.device_type == 'GPU']

get_available_gpus(), tf.config.list_physical_devices(
    device_type=None
)
```

```python
training_data=pd.read_csv('train.csv')
testing_data=pd.read_csv('test.csv')
validation_data=pd.read_csv('Dig-MNIST.csv')
sample_submission_data=pd.read_csv('sample_submission.csv')
```

```python
x_train=training_data.loc[:,"pixel0": ].values
y_train=training_data.loc[:, 'label'].values
x_test=testing_data.loc[:,"pixel0": ].values
x_test_id=testing_data['id'].values
x_valid=validation_data.loc[:, "pixel0":].values
y_valid=validation_data.loc[:, 'label'].values
```

```python
x_train.shape, x_test.shape, x_valid.shape
```

```python
x_train=x_train.reshape((60000, 28, 28, 1))/255
x_test=x_test.reshape((5000, 28, 28, 1))/255
x_valid=x_valid.reshape((10240, 28, 28, 1))/255
set(y_train), len(set(y_train))
```

```python
x_train.shape, x_test.shape, x_valid.shape
```

```python
batch_size=1000
n_classes=10
epochs=600
#you might want to decrease the epochs, because it converges at 200. and from 200 to 600, the accuracy improves from
#98.64 to 99.31.
```

```python
y_train=keras.utils.to_categorical(y_train)
y_valid=keras.utils.to_categorical(y_valid)
```

```python
model=Sequential()

model.add(Conv2D(16 ,kernel_size=(7, 7), padding='same', input_shape=(28, 28, 1), strides=(1, 1), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.1))

model.add(Conv2D(16 ,kernel_size=(7, 7), padding='same', strides=(1, 1), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.1))

model.add(Flatten())
model.add(Dense(32, activation='tanh'))
model.add(Dropout(0.1))

model.add(Dense(n_classes, activation='softmax'))#output_layer
```

```python
model.compile(loss='categorical_crossentropy', optimizer=keras.optimizers.SGD(), metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, batch_size=batch_size, epochs=epochs, validation_data=(x_valid, y_valid))
```

```python
predicted_classes=model.predict_classes(x_test)
sample_submission_data
```

```python
df=pd.DataFrame()
df['id']=x_test_id
df['label']=predicted_classes
```
